```python
# Importar paquetes necesarios
import numpy as np # Matrices
import pandas as pd # Trabajar con CSV
import matplotlib.pyplot as plt # Visualizar con datos
import io # Leer datos
from io import StringIO
```

```python
# Preparar variables para el bucle
temperatures = np.array([2000, 6000, 10000, 14000])
base_name = 'curvaDePlanck'

# Inicializar matrices vacas para almacenar los valores "x" e "y"
all_x = []
all_y = []

# Bucle a travs de cada temperatura
for temp in temperatures:
    # Construir el nombre del archivo CSV correspondiente
    filename = f"{base_name}_{temp}.csv"
    
    # Lectura del archivo CSV
    df = pd.read_csv(filename, sep=r"\s*,\s*", engine='python')
    
    # Almacenar los valores "x" e "y" del archivo
    x = df['x']
    y = df['y']
    all_x.append(x)
    all_y.append(y)

# Plotear todas las lneas en el mismo grfico
for i in range(len(all_x)):
    plt.plot(all_x[i], all_y[i], linewidth=2, label=f"{temperatures[i]} K")

# Aadir un ttulo y etiquetas para los ejes
plt.title("Distribucin de Planck", fontsize=15, pad=17)
plt.xlabel(r"Longitud de onda $\lambda$ (nm)", fontsize=12)
plt.ylabel(r"Radiacin $B_\lambda$ $(\frac{W}{m^2 \cdot sr \cdot nm})$", fontsize=12)

plt.legend()

#Limitar los ejes para mejor visualizacin
plt.xlim(0, 2e3)
plt.ylim(0, 2.5e15)

# Mostrar el grfico
plt.show()
```
